## Getting Started - Load Libraries and Dataset

```python
import quandl
import pandas as pd 
import numpy as np 
from datetime import datetime
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
quandl.ApiConfig.api_key = 'api key'
start = datetime(1990,1,1)
end = datetime(2010,1,1)

ticker = 'AAPL'

sns.set()
```

<a id='1'></a>
## Historical Price and Volume for 1 Stock

```python
data = quandl.get('WIKI/'+ticker,
              start_date=start,
              end_date=end)
data.head()
```

<a id='2'></a>
## Bullion Market

```python
df_gold = quandl.get('LBMA/GOLD',start_date=start,end_date=end)
df_gold
```

<a id='3'></a>
## Cryptocurrency

```python
# bitcoin price
btc = quandl.get('BCHAIN/MKPRU', start_date='2020-12-29', end_date='2021-12-29')
btc
```

<a id='4'></a>
## Unemployment Data

```python
unemployment = quandl.get('FRED/NROUST', start_date='2025-04-01', end_date='2031-10-01')
unemployment
```

<a id='5'></a>
## Mutual Funds

```python
# Mutual Fund Assets to GDP for World
mf = quandl.get('FRED/DDDI071WA156NWDB', start_date='1980-04-01', end_date='2020-10-01')
mf.plot(title = 'Mutual Fund Assets to GDP', figsize=(20, 6))
```

<a id='6'></a>
## Treasury Rates

```python
mf = quandl.get('USTREASURY/REALLONGTERM', start_date='2000-04-01', end_date='2020-10-01')
mf.plot(title = 'Treasury Real Long-Term Rates', figsize=(20, 6))
```
