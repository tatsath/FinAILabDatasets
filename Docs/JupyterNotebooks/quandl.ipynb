## Quandl
 Website to find Datasets: https://data.nasdaq.com/search
- Identifier is found under usage -> python -> examples and can then be put in the desiered functions, as shown below

```python
import quandl
import pandas as pd 
import numpy as np 
from datetime import datetime
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
quandl.ApiConfig.api_key = 'api key'
start = datetime(1990,1,1)
end = datetime(2010,1,1)

ticker = 'AAPL'

sns.set()
```

### Historical Data

```python
data = quandl.get('WIKI/'+ticker,
              start_date=start,
              end_date=end)
data.head()
```

### Bullion Market

```python
df_gold = quandl.get('LBMA/GOLD',start_date=start,end_date=end)
df_gold
```

### Cryptocurrency

```python
# bitcoin price
btc = quandl.get('BCHAIN/MKPRU', start_date='2020-12-29', end_date='2021-12-29')
btc
```

### Unemployment Data

```python
unemployment = quandl.get('FRED/NROUST', start_date='2025-04-01', end_date='2031-10-01')
unemployment
```

### Mutual Fund

```python
# Mutual Fund Assets to GDP for World
mf = quandl.get('FRED/DDDI071WA156NWDB', start_date='1980-04-01', end_date='2020-10-01')
mf.plot(title = 'Mutual Fund Assets to GDP', figsize=(20, 6))
```

### Treasury Rates

```python
mf = quandl.get('USTREASURY/REALLONGTERM', start_date='2000-04-01', end_date='2020-10-01')
mf.plot(title = 'Treasury Real Long-Term Rates', figsize=(20, 6))
```

### S&P Fundamentals

```python
sp = quandl.get('YALE/SPCOMP', start_date='2015-04-01', end_date='2021-10-01')
sp
```
