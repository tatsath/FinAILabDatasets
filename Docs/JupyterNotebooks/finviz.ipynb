## Getting Started - Load Libraries and Dataset

```python
from finvizfinance.quote import finvizfinance
import pandas as pd

stock = finvizfinance('tsla')
```

<a id='1'></a>
## Stock Fundamentals


```python
chart = stock.ticker_charts()
chart
```

```python
stock_fundament = stock.ticker_fundament()
# stock_fundament = pd.DataFrame(stock_fundament, index = ['Values']).T
stock_fundament
```

<a id='2'></a>
## Ticker Description

```python
description = stock.ticker_description()
description
```

<a id='3'></a>
## Ratings/News

```python
outer_ratings_df = stock.ticker_outer_ratings()
outer_ratings_df.head()
```

```python
news_df = stock.ticker_news()
news_df
```

```python
from finvizfinance.news import News

fnews = News()
all_news = fnews.get_news()

all_news['news'].head() # 'blogs'
```

<a id='4'></a>
## Insider Trades

```python
inside_trader_df = stock.ticker_inside_trader()
inside_trader_df.tail()
```

```python
from finvizfinance.insider import Insider

finsider = Insider(option='top owner trade')
# option: latest, top week, top owner trade
# default: latest

insider_trader = finsider.get_insider()
insider_trader.head()
```

## Multiple Ticker Overview

```python
from finvizfinance.screener.overview import Overview

foverview = Overview()
filters_dict = {'Index':'S&P 500','Sector':'Basic Materials'}
foverview.set_filter(filters_dict=filters_dict)
df = foverview.screener_view()
df.head()
```
